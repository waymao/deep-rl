```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_figheight(4)
fig.set_figwidth(4)
ax.set_xlim((0, 10))
ax.set_ylim((0, 10))

w = 0.5
h = 0.5

for x in np.arange(0, 10, .5):
    for y in np.arange(0, 10, .5):
        if (x + .25 - 3) ** 2 + (y + .25 - 3) ** 2 <= 2 ** 2:
            ax.add_patch(patches.Rectangle((x, y), w, h, fill=False, edgecolor='grey', facecolor='yellow'))
        else:
            ax.add_patch(patches.Rectangle((x, y), w, h, fill=False, edgecolor='grey', facecolor='green'))
ax.add_patch(patches.Circle(xy=(3, 3), radius=2, edgecolor="red", fill=False))
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_figheight(4)
fig.set_figwidth(4)
ax.set_xlim((0, 10))
ax.set_ylim((0, 10))

n_dots = 120   # set number of dots

xs, ys = [], []    # for coordinates of points to plot
xs_good, ys_good = [], []

for x in np.arange(.25, 10, .5):
    for y in np.arange(.25, 10, .5):
        if (x - 3) ** 2 + (y - 3) ** 2 <= 2 ** 2:
            xs.append(x)   # collect x
            ys.append(y)   # collect y
        else:
            xs_good.append(x)
            ys_good.append(y)

ax.add_patch(patches.Circle(xy=(3, 3), radius=2, edgecolor="red", fill=False))
ax.scatter(xs, ys, c = 'orange', s=5, marker="X")  # plot points 
ax.scatter(xs_good, ys_good, c = 'green', s=5, marker="X")  # plot points 
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_figheight(4)
fig.set_figwidth(4)
ax.set_xlim((0, 10))
ax.set_ylim((0, 10))

n_dots = 120   # set number of dots

xs, ys = [], []    # for coordinates of points to plot
xs_good, ys_good = [], []

sampled_points = np.random.normal(loc=(3, 3), scale=2, size=(400, 2))

for (x, y) in sampled_points:
        if (x - 3) ** 2 + (y - 3) ** 2 <= 2 ** 2:
            xs.append(x)   # collect x
            ys.append(y)   # collect y
        else:
            xs_good.append(x)
            ys_good.append(y)

ax.add_patch(patches.Circle(xy=(3, 3), radius=2, edgecolor="red", fill=False))
ax.scatter(xs, ys, c = 'orange', s=5, marker="X", label="fail")  # plot points 
ax.scatter(xs_good, ys_good, c = 'green', s=5, marker="X", label="success")  # plot points 
# ax.legend()
plt.show()
```
